```python
#import ollama
import requests
import json
```

```python
url = "http://127.0.0.1:11434/api/chat"  ## You can check other endpoints ##

# Define prompt
myprompt = {
    "model": "llama3.1",
    "messages": [{"role": "user", "content": "What is banco del bajio?"}]
}

response = requests.post(url, json=myprompt, stream=True)

if response.status_code == 200:
    print("Streamming response from Ollama:")
    for line in response.iter_lines(decode_unicode=True):
        if line:
            try:
                json_data = json.loads(line)
                if "message" in json_data and "content" in json_data["message"]:
                    print(json_data["message"]["content"], end="")
            except json.JSONDecodeError:
                print(f"\nFailed to parse line: {line}")
#else:
```
